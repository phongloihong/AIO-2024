### Get movies have revenue > 5% total of dataset with rating < 6 from 2010-2015

```python
df[(df['Rating'] < 6) & (df['Year'] >= 2010) & (df['Year'] < 2016) & (df['Revenue (Millions)'] > df['Revenue (Millions)'].quantile(0.9))]
```

### Group by director rating

```python
df.groupby('Director')['Rating'].mean().sort_values(ascending=False).head()
```

## Due with missing values

```python
df.isnull().sum().sort_values(ascending=False)

remove_missing_data = df.dropna(inplace=False)
remove_missing_data.isnull().sum().sort_values(ascending=False)

revenue_mean = df['Revenue (Millions)'].mean()
df['Revenue (Millions)'] = df['Revenue (Millions)'].fillna(revenue_mean)
mean_metascore = df['Metascore'].mean()
df['Metascore'] = df['Metascore'].fillna(mean_metascore)
df.isnull().sum().sort_values(ascending=False)
```

### Apply rating value

```python
def rating_category(rating):
    if rating >= 8.0:
        return 'Good'
    elif rating >= 6.0:
        return 'Average'
    else:
        return 'Bad'

df['Rating Category'] = df['Rating'].apply(rating_category)
```
